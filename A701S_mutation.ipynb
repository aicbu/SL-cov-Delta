```python
import pandas as pd
from Bio import SeqIO
```

```python
met=pd.read_csv("data_generation/samples.csv", parse_dates=['date'])
#met=metadata[['strain','date','division']]drop
len(met)
```

```python
id=[]
mut701=[]
mut222=[]
mut1078=[]
for record in SeqIO.parse("data_generation/input_seqs.gene.S.fasta", "fasta"):
    id.append(record.id)
    mut701.append(record.seq[700])
    mut222.append(record.seq[221])
    mut1078.append(record.seq[1077])
```

```python
len(id)
```

```python
dict={'seq_id':id,'pos_222':mut222,'pos_701':mut701,'pos_1078':mut1078}
df=pd.DataFrame(dict)
df["seq_id"] = df["seq_id"].str.replace('hCoV-19/SriLanka/','',regex=True)
df["seq_id"] = df["seq_id"].str.replace('/2021','',regex=True)
df
```

```python
print('No of N at pos222 ='+str(len(df[df.pos_222=='X'])))
print('No of A at pos222 ='+str(len(df[df.pos_222=='A'])))
print('No of V at pos222 ='+str(len(df[df.pos_222=='V'])))
```

```python
print('No of N at pos701 ='+str(len(df[df.pos_701=='X'])))
print('No of A at pos701 ='+str(len(df[df.pos_701=='A'])))
print('No of S at pos701 ='+str(len(df[df.pos_701=='S'])))
```

```python
print('No of N at pos1078 ='+str(len(df[df.pos_1078=='X'])))
print('No of A at pos1078='+str(len(df[df.pos_1078=='A'])))
print('No of S at pos1078='+str(len(df[df.pos_1078=='S'])))
```

```python
merged=pd.merge(left=df, right=met, how='outer', left_on='seq_id', right_on='sample')
merged.drop('Pateints Name', 1, inplace=True)
merged
```

```python
merged.set_index('date', inplace=True)
```
